# Finding minima using gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

## The function and its gradient

$ y = -100 x^2 - x^3 + x^4$

$\frac{dy}{dx} = -200 x - 3 x^2 + 4 x^3$

```python
NumberGrid = 100

x = np.linspace(-10,10,NumberGrid)
y = np.zeros( NumberGrid )

for i in range(NumberGrid) :
    y[i] = -100*x[i]**2 - x[i]**3 + x[i]**4

plt.plot( x , y )
plt.show()
```

## Performing gradient descent

$ x \rightarrow x - \eta \; \frac{dy}{dx} $

Parameters

```python
# Initial value
x0 = 2.5

# Hyperparameters
MaxIterations = 200
eta = 0.0001
```

```python
xi = x0
xHistory = np.zeros( MaxIterations )

for i in range( MaxIterations ):
    dydx = -200*xi - 3*xi**2 + 4*xi**3

    xi = xi - eta * (dydx)
    xHistory[i] = xi

plt.plot( x , y )
plt.plot( x0 , -100*x0**2 - x0**3 + x0**4 , '*b' )
plt.plot( xi , -100*xi**2 - xi**3 + xi**4 , 'or' )
plt.legend( ['Function','Initial value','Final value'] )
plt.title( "Initial and final value" )
plt.xlabel( "x" )
plt.ylabel( "y" )
plt.show()

plt.plot( xHistory,'-o' )
plt.title( "Evolution of the solution" )
plt.xlabel( "Iteration" )
plt.ylabel( "x value" )
plt.show()
```
