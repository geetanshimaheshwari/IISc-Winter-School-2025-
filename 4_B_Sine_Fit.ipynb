## Code to fit a function to a data using Stochastic Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Input parameters:   $(A,m,\phi,c)$

<div style="border: 2px solid #535353; width: 820px; padding: 10px; background-color: #f2f2f2; border-radius: 8px;">
<div style="background-color: white; padding: 10px; border-radius: 5px; max-width: 800px; overflow: auto;">

 $$
 y = A \; \sin(m \: x + \phi) + c
 $$

 </div>
 </div>

```python
A = 0.5        # Slope
m = 1          # Intercept
phi = np.pi/3  # Phase
c = 1.0        # Vertical shift
```

### Data generation:

```python
L = 2*np.pi      # Range of x is [0,L]
Np = 200     # Number of data points
```

```python
x = L * np.random.rand(Np)
y = A * np.sin(m * x + phi) + c
```

```python
plt.plot(x,y,'*')
plt.show()
```

### Add noise to data: $ y \rightarrow y + \xi $

```python
xi = 0.01 * np.random.randn(Np)

plt.hist(xi, bins=10, color='skyblue', edgecolor='black')
plt.show()
```

```python
y = y + xi
plt.plot(x,y,'*')
plt.show()
```

### Initial guess: $(m_0,c_0)$

```python
A0 = 0.5
m0 = 1.5
phi0 = 0 #np.pi/2.0
c0 = 0 #0.5
```

### Loss function:

<div style="border: 2px solid #535353; width: 820px; padding: 10px; background-color: #f2f2f2; border-radius: 8px;">
<div style="background-color: white; padding: 10px; border-radius: 5px; max-width: 800px; overflow: auto;">

$$
E(A,m,\phi,c) = \frac{1}{N_p} \sum_{i=1}^{N_p} [y_i - ( A \; \sin(m \: x_i + \phi) + c)]^2 
$$

</div>
</div>

### Gradient:

<div style="border: 2px solid #535353; width: 820px; padding: 10px; background-color: #f2f2f2; border-radius: 8px;">
<div style="background-color: white; padding: 10px; border-radius: 5px; max-width: 800px; overflow: auto;">

$$ 
E(A,m,\phi,c) = \frac{1}{N_p} \sum_{i=1}^{N_p} (y_i - ( A \; \sin(m \: x + \phi) + c))^2 
$$
$$
\frac{\partial E}{\partial A} = - \frac{2}{N_p} \sum_{i=1}^{N_p} (y_i-(A \sin(m x_i + \phi)+c)) \; \sin(m x_i + \phi)
$$
$$
\frac{\partial E}{\partial m} = - \frac{2}{N_p} \sum_{i=1}^{N_p} A \; (y_i-(A \sin(m x_i + \phi)+c)) \; \cos(m x_i + \phi) \; x_i
$$
$$
\frac{\partial E}{\partial \phi} = - \frac{2}{N_p} \sum_{i=1}^{N_p} A \; (y_i-(A \sin(m x_i + \phi)+c)) \; \cos(m x_i + \phi)
$$
$$
\frac{\partial E}{\partial c} = - \frac{2}{N_p} \sum_{i=1}^{N_p} (y_i-(A \sin(m x + \phi)+c))
$$

</div>
</div>

### Gradient descent:

<div style="border: 2px solid #535353; width: 420px; padding: 10px; background-color: #f2f2f2; border-radius: 8px;">
<div style="background-color: white; padding: 10px; border-radius: 5px; max-width: 400px; overflow: auto;">

$$
\begin{align*}
A \; & \rightarrow & A - \eta \: \frac{\partial E}{\partial A} \\
m \; & \rightarrow & m - \eta \: \frac{\partial E}{\partial m} \\
\phi \; & \rightarrow & \phi - \eta \: \frac{\partial E}{\partial \phi} \\
c \; & \rightarrow & c - \eta \: \frac{\partial E}{\partial c}
\end{align*}
$$

</div>
</div>

```python
eta = 0.01  # Learning rate
n_iter = 20000
```

## Using Gradient Descent:

```python
A_curr = A0
m_curr = m0
phi_curr = phi0
c_curr = c0

AA = np.zeros(n_iter)
mm = np.zeros(n_iter)
pp = np.zeros(n_iter)
cc = np.zeros(n_iter)
EE = np.zeros(n_iter)

for it in range(n_iter):

    yy = y
    xx = x

    AA[it] = A_curr
    mm[it] = m_curr
    pp[it] = phi_curr
    cc[it] = c_curr
    EE[it] = np.sum((yy - (A_curr * np.sin(m_curr * xx + phi_curr) + c_curr))**2) / Np

    dE_dA = 0
    dE_dm = 0
    dE_dphi = 0
    dE_dc = 0

    for i in range(len(xx)):
        dE_dA += -2 * np.sin(m_curr * xx[i] + phi_curr) * (yy[i] - (A_curr * np.sin(m_curr * xx[i] + phi_curr) + c_curr)) / Np
        dE_dm += -2 * A_curr * xx[i] * np.cos(m_curr * xx[i] + phi_curr) * (yy[i] - (A_curr * np.sin(m_curr * xx[i] + phi_curr) + c_curr)) / Np
        dE_dphi += -2 * A_curr * np.cos(m_curr * xx[i] + phi_curr) * (yy[i] - (A_curr * np.sin(m_curr * xx[i] + phi_curr) + c_curr)) / Np
        dE_dc += -2 * (yy[i] - (A_curr * np.sin(m_curr * xx[i] + phi_curr) + c_curr)) / Np

    A_curr = A_curr - eta * dE_dA
    m_curr = m_curr - eta * dE_dm
    phi_curr = phi_curr - eta * dE_dphi
    c_curr = c_curr - eta * dE_dc
```

```python
EE_GD = EE.copy()
```

## Using Stochastic Gradient Descent:

```python
A_curr = A0
m_curr = m0
phi_curr = phi0
c_curr = c0

AA = np.zeros(n_iter)
mm = np.zeros(n_iter)
pp = np.zeros(n_iter)
cc = np.zeros(n_iter)
EE = np.zeros(n_iter)

batch_size = 10
r = np.int64(Np / batch_size)

for it in range(n_iter): 

    AA[it] = A_curr
    mm[it] = m_curr
    pp[it] = phi_curr
    cc[it] = c_curr
    EE[it] = np.sum((yy - (A_curr * np.sin(m_curr * xx + phi_curr) + c_curr))**2) / Np

    for jj in range(r):
        ii = np.int64(np.floor( np.random.rand(batch_size) * Np ))

        yy = y[ii]
        xx = x[ii]

        dE_dA = 0
        dE_dm = 0
        dE_dphi = 0
        dE_dc = 0

        for i in range(len(xx)):
            dE_dA += -2 * np.sin(m_curr * xx[i] + phi_curr) * (yy[i] - (A_curr * np.sin(m_curr * xx[i] + phi_curr) + c_curr)) / Np
            dE_dm += -2 * A_curr * xx[i] * np.cos(m_curr * xx[i] + phi_curr) * (yy[i] - (A_curr * np.sin(m_curr * xx[i] + phi_curr) + c_curr)) / Np
            dE_dphi += -2 * A_curr * np.cos(m_curr * xx[i] + phi_curr) * (yy[i] - (A_curr * np.sin(m_curr * xx[i] + phi_curr) + c_curr)) / Np
            dE_dc += -2 * (yy[i] - (A_curr * np.sin(m_curr * xx[i] + phi_curr) + c_curr)) / Np

        A_curr = A_curr - eta * dE_dA
        m_curr = m_curr - eta * dE_dm
        phi_curr = phi_curr - eta * dE_dphi
        c_curr = c_curr - eta * dE_dc
```

```python
EE_SGD = EE.copy()
```

### Compare the results:

### The solution:

```python
plt.plot(x,y,'*', label='Data points')
y_fit = A_curr * np.sin(m_curr * x + phi_curr) + c_curr
plt.plot(x, y_fit,'.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
