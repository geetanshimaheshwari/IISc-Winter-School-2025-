## Code to fit a straight line to a data using Stochastic Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Input parameters:   $(m,c)$
 $y = m x + c$

```python
m = 0.5    # Slope
c = 4      # Intercept
```

### Data generation: $ y = m x + c $

```python
L = 10      # Range of x is [0,L]
Np = 50     # Number of data points
```

```python
x = L * np.random.rand(Np)
y = m * x + c
```

```python
plt.plot(x,y,'*')
plt.show()
```

### Add noise to data: $ y \rightarrow y + \xi $

```python
xi = 0.3 * np.random.randn(Np)

plt.hist(xi, bins=10, color='skyblue', edgecolor='black')
```

```python
y = y + xi
plt.plot(x,y,'*')
```

### Initial guess: $(m_0,c_0)$

```python
m0 = 0
c0 = 0
```

### Loss function:

$ E(m,c) = \frac{1}{N_p} \sum_{i=1}^{N_p} (y_i - (m \; x_i + c))^2 $

```python
xmin = 0
xmax = L

y1 = m * xmin + c
y2 = m * xmax + c

ymin = np.min([y1,y2])
ymax = np.max([y1,y2])

N_grid = 20

m_grid = np.linspace(-5*m,5*2*m,N_grid)
c_grid = np.linspace(c-30,c+30,N_grid)

E = np.zeros((N_grid,N_grid))

for i in range(N_grid):
    for j in range(N_grid):
        for k in range(Np):
            mm = m_grid[i]
            cc = c_grid[j]
            E[i,j] += (y[k] - (mm * x[k] + cc))**2 / Np

plt.contour(c_grid, m_grid, E, levels=50, cmap='viridis')
plt.colorbar()
plt.plot(c, m, 'r*', markersize=12)
plt.xlabel('c')
plt.ylabel('m')
plt.title('Loss function E(m,c)')
plt.show()
```

### Gradient:

$$
\frac{\partial E}{\partial m} = -\frac{2}{N_p} \sum_{i=1}^{N_p} x_i \: \left(y_i- (m \: x_i + c)\right) 
$$

$$
\frac{\partial E}{\partial c} = -\frac{2}{N_p} \sum_{i=1}^{N_p} (y_i- (m \: x_i + c)) 
$$

### Gradient descent:

$$
\begin{align*}
m \; & \rightarrow & m - \eta \: \frac{\partial E}{\partial m} \\
c \; & \rightarrow & c - \eta \: \frac{\partial E}{\partial c}
\end{align*}
$$

```python
eta = 0.001  # Learning rate
n_iter = 1000

m_curr = m0
c_curr = c0

mm = np.zeros(n_iter)
cc = np.zeros(n_iter)
EE = np.zeros(n_iter)

for it in range(n_iter):
    mm[it] = m_curr
    cc[it] = c_curr
    EE[it] = np.sum((y - (m_curr * x + c_curr))**2) / Np

    dE_dm = 0
    dE_dc = 0
    for i in range(Np):
        dE_dm += -2 * x[i] * (y[i] - (m_curr * x[i] + c_curr)) / Np
        dE_dc += -2 * (y[i] - (m_curr * x[i] + c_curr)) / Np
    
    m_curr = m_curr - eta * dE_dm
    c_curr = c_curr - 10 * eta * dE_dc
```

```python
EE_GD = EE.copy()
```

## Using Stochastic Gradient Descent:

```python
eta = 0.001  # Learning rate
n_iter = 1000

m_curr = m0
c_curr = c0

mm = np.zeros(n_iter)
cc = np.zeros(n_iter)
EE = np.zeros(n_iter)

batch_size = 10
r = np.int64(Np / batch_size)

for it in range(n_iter):
    mm[it] = m_curr
    cc[it] = c_curr
    EE[it] = np.sum((y - (m_curr * x + c_curr))**2) / Np

    for jj in range(r):
        ii = np.int64(np.floor( np.random.rand(batch_size) * Np ))
        yy = y[ii]
        xx = x[ii]

        dE_dm = 0
        dE_dc = 0

        for i in range(batch_size):
            dE_dm += -2 * xx[i] * (yy[i] - (m_curr * xx[i] + c_curr)) / batch_size
            dE_dc += -2 * (yy[i] - (m_curr * xx[i] + c_curr)) / batch_size
        
        m_curr = m_curr - eta * dE_dm
        c_curr = c_curr - 10 * eta * dE_dc
```

```python
EE_SGD = EE.copy()
```

```python
m_curr,c_curr,m,c
```

### The solution:

```python
plt.plot(x,y,'*', label='Data points')
y_fit = m_curr * x + c_curr
plt.plot(x, y_fit, 'r-', label='Fitted line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

### Generate a GIF of the line fit's evolution

```python
# Generate a GIF of the line fit's evolution
import imageio
import imageio.v3 as iio

images = []

for i in range(0, n_iter, n_iter//10):
    plt.plot(x,y,'*', label='Data points')
    y_fit = m * x + c
    plt.plot(x, y_fit, 'y--', label='True line')
    y_fit = mm[i] * x + cc[i]
    plt.plot(x, y_fit, 'r-', label=f'Iter {i}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Linear Regression Fit')
    plt.legend()
    
    filename = 'Fig/tmp.png'
    plt.savefig(filename)
    images.append(imageio.imread(filename))
    plt.close()

iio.imwrite(
    'Fig/linreg_evolution.gif',
    images,
    duration=200,  # Duration of each frame in milliseconds (e.g., 200ms = 0.2 seconds)
    loop=0         # 0 means the GIF will loop indefinitely
)
```
