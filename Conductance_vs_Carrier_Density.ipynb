The conductance of a two-dimensional electronic system increases with carrier density due to the proportionality between conductivity and charge carrier concentration. In the diffusive transport regime, the conductivity is given by σ=neμ, where n is the carrier density and μ is the mobility. Assuming weak density dependence of mobility, conductance shows an approximately linear dependence on density. Deviations from linearity indicate changes in scattering mechanisms or mobility variation with carrier concentration.

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data1 = np.loadtxt("Density(m^-2) vs Conductance(S).txt")
data2 = np.loadtxt("Density(m^-2) vs Conductance(S) 2.txt")
data3 = np.loadtxt("Density(m^-2) vs Conductance(S) 3.txt")

# Separate columns
n1, G1 = data1[:, 0], data1[:, 1]
n2, G2 = data2[:, 0], data2[:, 1]
n3, G3 = data3[:, 0], data3[:, 1]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(n1, G1, label="Dataset 1")
plt.plot(n2, G2, label="Dataset 2")
plt.plot(n3, G3, label="Dataset 3")

# Labels
plt.xlabel("Carrier Density (m$^{-2}$)")
plt.ylabel("Conductance (S)")
plt.title("Conductance vs Carrier Density")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.plot(n1, G1 / np.max(G1), label="Dataset 1 (normalized)")
plt.plot(n2, G2 / np.max(G2), label="Dataset 2 (normalized)")
plt.plot(n3, G3 / np.max(G3), label="Dataset 3 (normalized)")

plt.xlabel("Carrier Density (m$^{-2}$)")
plt.ylabel("Normalized Conductance")
plt.title("Normalized Conductance vs Density")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
