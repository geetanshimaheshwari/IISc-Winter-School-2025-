# Let us import useful packages

```
# some standard python imports
import numpy as np
import copy
import matplotlib.pyplot as plt

# make sure lenstronomy is installed, otherwise install the latest pip version
try:
    import lenstronomy
except:
    !pip install lenstronomy
```

# Let us see what a point mass lens gives

```
# import the LensModel class #
from lenstronomy.Plots import lens_plot
from lenstronomy.LensModel.lens_model import LensModel

# specify the choice of lens models #
lens_model_list = ['POINT_MASS']

point_1 = {'theta_E':1.0, "center_x": 0.0, "center_y": 0.0}

kwargs_lens = [point_1]

lensModel = LensModel(lens_model_list=lens_model_list, z_lens=0.5, z_source=1.5)

# image plane coordinate #
beta_ra, beta_dec = 0.1, 0.0

f, axex = plt.subplots(1, 1, figsize=(5, 5), sharex=False, sharey=False)
lens_plot.lens_model_plot(axex, lensModel=lensModel, kwargs_lens=kwargs_lens, sourcePos_x=beta_ra, sourcePos_y=beta_dec, point_source=True, with_caustics=True, fast_caustic=True, coord_inverse=False)
axex.scatter(0, 0, marker="+", color="red")

plt.show()
```

# Let us use a little more complicated lens model made of two point mass lenses

```
# import the LensModel class #
from lenstronomy.LensModel.lens_model import LensModel

# specify the choice of lens models #
lens_model_list = ['POINT_MASS', 'POINT_MASS']

point_1 = {'theta_E':1.0, "center_x": 0.0, "center_y": 0.0}
point_2 = {'theta_E':1.0, "center_x": 1.0, "center_y": 1.0}

kwargs_lens = [point_1, point_2]

lensModel = LensModel(lens_model_list=lens_model_list, z_lens=0.5, z_source=1.5)

# image plane coordinate #
beta_ra, beta_dec = 0.5, 0.5

f, axex = plt.subplots(1, 1, figsize=(5, 5), sharex=False, sharey=False)
lens_plot.lens_model_plot(axex, lensModel=lensModel, kwargs_lens=kwargs_lens, sourcePos_x=beta_ra, sourcePos_y=beta_dec, point_source=True, with_caustics=True, fast_caustic=True, coord_inverse=False)
axex.scatter(0, 0, marker="+", color="red")
axex.scatter(1, 1, marker="+", color="red")

plt.show()
```
